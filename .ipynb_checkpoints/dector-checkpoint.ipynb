# Import Packages

```python
import numpy as np
import pandas as pd
import nltk
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import scikitplot as skplt
```

```python
df = pd.read_csv('./data/spam1.csv', encoding='latin-1')
df= df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'])
df = df.rename(columns = {'v1' : 'label', 'v2' : 'text' })
```

```python
# get labels
dataLabels = df.label
```

# Data Visualization

```python
stopwords= list(STOPWORDS) + nltk.corpus.stopwords.words('english')


ham_dataset=df[df.label == 'ham']
spam_dataset=df[df.label == 'spam']
ham_words = ''
spam_words=''

for words in ham_dataset.text:
    txt = words.lower()
    tokens = nltk.word_tokenize(txt)
    for word in tokens:
        ham_words += word + " "
for words in spam_dataset.text:
    txt = words.lower()
    tokens = nltk.word_tokenize(txt)
    for word in tokens:
        spam_words += word + " "
def plot_wordcloud(wordcloud):
    plt.figure( figsize=(10,8), facecolor='k')
    plt.imshow(wordcloud)
    plt.tight_layout(pad=0)
    plt.axis('off')
    plt.show()
```

## ham words

```python
hamWordCloud =WordCloud(background_color='white',width=500, height=500,stopwords=stopwords,max_words=500,max_font_size=50).generate(ham_words)
plot_wordcloud(hamWordCloud)
```

## Spam WordCloud

```python
spamWordCloud =WordCloud(background_color='black',width=500, height=500,stopwords=stopwords,max_words=500,max_font_size=50).generate(ham_words)
plot_wordcloud(spamWordCloud)
```

```python
count_Class=pd.value_counts(df.label, sort= True)

# Data to plot
labels = 'Ham', 'Spam'
sizes = [count_Class[0], count_Class[1]]
colors = ['green', 'red']
explode = (0.1, 0.1)  # explode  spam
 
# Plot
plt.pie(sizes, labels=labels, colors = colors, explode = explode,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.show()
```

# Split Data

```python
X_train, X_test, y_train, y_test = train_test_split(df.text, dataLabels,test_size=0.3, random_state=42)
```

```python
# vectorize the text data
predictions= dict()
cvec=CountVectorizer(stop_words='english', ngram_range=(1,2), max_features=100)
stemmer = nltk.stem.snowball.SnowballStemmer('english')
lemmer= nltk.stem.WordNetLemmatizer()
cvec.fit(X_train)
X_train = cvec.transform(X_train)
X_test= cvec.transform(X_test)
```

# Test Different Models

## Naive Bayes

```python
clf=MultinomialNB()
clf.fit(X_train, y_train)
predicted_values = clf.predict(X_test)
acurracy = accuracy_score(y_test,predicted_values)
predictions['Naive Bayes'] = acurracy * 100
confusionMatrix = confusion_matrix(y_test, predicted_values)
print(f"The accuracy of the Naive Bayes model is {acurracy * 100} %")
print(confusionMatrix)
skplt.metrics.plot_confusion_matrix(y_test,predicted_values, normalize=True)
plt.show()


```

## Decision Tree

```python
DT=DecisionTreeClassifier()
DT.fit(X_train, y_train)
predicted_values_DT = DT.predict(X_test)
acurracy_DT = accuracy_score(y_test,predicted_values_DT)
predictions['DecisionTree']= acurracy_DT*100
print(f"The accuracy of the Decision Tree model  is {acurracy_DT*100 }%")
confusion_matrix_DT = confusion_matrix(y_test,predicted_values_DT)
print(confusion_matrix_DT)
skplt.metrics.plot_confusion_matrix(y_test,predicted_values_DT, normalize=True)
plt.show()
```

 ## Gradient Boosted Tree

```python
clf = GradientBoostingClassifier(random_state=0)
clf.fit(X_train, y_train)
predicted_values_GBT = clf.predict(X_test)
acurracy_GBT = accuracy_score(y_test,predicted_values_GBT)
predictions['GradientBoostedTree'] = acurracy_GBT * 100
print(f"The accuracy of the Gradient Boosted Tree model  is {acurracy_DT*100 }%")
confusion_matrix_GBT = confusion_matrix(y_test,predicted_values_GBT)
print(confusion_matrix_GBT)
skplt.metrics.plot_confusion_matrix(y_test,predicted_values_GBT, normalize=True)
plt.show()
```
